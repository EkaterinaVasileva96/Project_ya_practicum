```python
#Найдем медианные значения дохода по каждому типу занятости

employee_median_total_income = data[data['income_type'] == 'сотрудник']['total_income'].median()
companion_median_total_income = data[data['income_type'] == 'компаньон']['total_income'].median()
retire_median_total_income = data[data['income_type'] == 'пенсионер']['total_income'].median()
governer_median_total_income = data[data['income_type'] == 'госслужащий']['total_income'].median()
business_median_total_income = data[data['income_type'] == 'предприниматель ']['total_income'].median()
student_median_total_income = data[data['income_type'] == 'студент']['total_income'].median()
decree_median_total_income = data[data['income_type'] == 'в декрете']['total_income'].median()

print('Медиана ежемесячного дохода сотрудников составляет', employee_median_total_income)
print('Медиана ежемесячного дохода компаньонов составляет', companion_median_total_income)
print('Медиана ежемесячного дохода пенсионеров составляет', retire_median_total_income)
print('Медиана ежемесячного дохода госслужащих составляет', governer_median_total_income)
print('Медиана ежемесячного дохода предпринимателей составляет', business_median_total_income)
print('Медиана ежемесячного дохода студентов составляет', student_median_total_income)
print('Медиана ежемесячного дохода в декрете составляет', decree_median_total_income)
```

```python
# Заполним пропущенные значения
data.loc[data['income_type'] == 'сотрудник', 'total_income']=data.loc[data['income_type'] == 'сотрудник', 'total_income'].fillna(value=employee_median_total_income)

data.loc[data['income_type'] == 'компаньон', 'total_income']=data.loc[data['income_type'] == 'компаньон', 'total_income'].fillna(value=companion_median_total_income)

data.loc[data['income_type'] == 'пенсионер', 'total_income']=data.loc[data['income_type'] == 'пенсионер', 'total_income'].fillna(value=retire_median_total_income)

data.loc[data['income_type'] == 'госслужащий', 'total_income']=data.loc[data['income_type'] == 'госслужащий', 'total_income'].fillna(value=governer_median_total_income)

data.loc[data['income_type'] == 'студент', 'total_income']=data.loc[data['income_type'] == 'студент', 'total_income'].fillna(value=student_median_total_income)

data.loc[data['income_type'] == 'в декрете', 'total_income']=data.loc[data['income_type'] == 'в декрете', 'total_income'].fillna(value=decree_median_total_income)

# Медианное значение дохода предпринимателей не посчитался, возможно по ним отсутствуют данные. Можно удалить эти строки
#так как их всего две и они не повлияют на значения результатов

display(data.head(50))

```

```python
#найдем ср. знач  стажа 
days_employed_mean = data['days_employed'].mean()

#заменим отрицательные числа на положительные в стаже
def negative_to_pozitive(value):
    if value < 0:
        value *= -1
        return value
    else:
        return value
data['days_employed'] = data['days_employed'].apply(negative_to_pozitive)
data['days_employed'] = data['days_employed'].fillna(days_employed_mean)



display(data.head(30))
```

### Лемматизация

```python
# рассмотрим столбец purpose
data['purpose'].value_counts()
```

```python
# лемматизируем столбец "purpose"
# для этого определим функцию row_lemma(purpose)

from pymystem3 import Mystem
m = Mystem()

def row_lemma(purpose):
    lemma = ' '.join(m.lemmatize(purpose))
    return lemma

# создадим столбец purpose_lem, в котором будут леммы столбца purpose
data['purpose_lem'] = data['purpose'].apply(row_lemma)

# посчитаем количество уникальных лемм и выведем их в порядке убывания:
from collections import Counter
purpose_lem_count = Counter(data['purpose_lem'])
purpose_lem_count.most_common()
```

**Вывод**

В данном случае лемматизация удобнее стемминга, так как заемщики использовали целые словосочетания для описания цели. В дальнейшем после лемматизации будет удобнее пользоваться отбором по столбцу "purpose", выбирая словарную форму слова для поиска.

### Категоризация данных

```python
# создадим столбец purpose_category на основании данных лемматизации столбца purpose_lem
def purpose_category(purpose_lem):
    if 'автомобиль' in purpose_lem:
        return "сделки с автомобилем"
    if "образование" in purpose_lem:
        return "образование"
    if "свадьба" in purpose_lem:
        return "свадьба"
    if "недвижимость" in purpose_lem or "строительство" in purpose_lem or "жилье" in purpose_lem:
        return "сделки с недвижимостью"
   
data['purpose_category'] = data['purpose_lem'].apply(purpose_category)

data['purpose_category'].value_counts()
```

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
#Посчитаем процент заемщиков с задолженностью по возврату кредита

debt_per = data['debt'][(data['debt'] == 1)].count() / data['debt'].count()
print('{:.2%}'.format(debt_per), 'среди заемщиков не возвращали кредит вовремя')
```

```python
# Построим сводную таблицу на основании исходной обработанной используя столбцы "children", "debt" и "family_status"

data_pivot_children = data.pivot_table(index=['children'], columns='debt', values = 'family_status', aggfunc='count')

# Для удобства обработки информации переименуем слобцы
names_columns = ['Добросовестный плательщик', 'Должник']
data_pivot_children.set_axis(names_columns, axis = "columns", inplace = True)

# Добавим еще столбцы с расчетом доли должников с разным количеством детей и столбец с отклонением от общей доли должников

data_pivot_children['Доля должников'] = data_pivot_children['Должник'] / data_pivot_children['Добросовестный плательщик']
data_pivot_children['Отклонение от общего уровня должников'] = data_pivot_children['Доля должников'] - debt_per
display(data_pivot_children)
```

**Вывод**

На основании сводной таблицы наблюдается, что задолженности по кредитам увеличивается прямопропорционально появлению в семье детей.
